```python
import numpy as np
import pandas as pd
```

```python
subjects = pd.read_csv('data/subjects.csv', index_col=0)
NASA_TLI = pd.read_csv('data/NASA_TaskLoadIndex.csv', index_col=0).T
SVM_CrossVal = pd.read_csv('data/SVM_CrossVal.csv', index_col=0)
CNN_CrossVal = pd.read_csv('data/CNN_CrossVal.csv', index_col=0)
```

```python
# Convert the gender column to a categorical variable
gender = {'male': 1,'female': 0}
subjects.Sex = [gender[item] for item in subjects.Sex]
```

```python
# From the nature of the NASA TLI 
# for some questions lower values are better
# so we need to reverse the scale for those questions
NASA_TLI['Mental Demand'] = 20 - NASA_TLI['Mental Demand']
NASA_TLI['Physical Demand'] = 20 - NASA_TLI['Physical Demand']
NASA_TLI['Temporal Demand'] = 20 - NASA_TLI['Temporal Demand']
NASA_TLI['Effort'] = 20 - NASA_TLI['Effort']
NASA_TLI['Frustration'] = 20 - NASA_TLI['Frustration']
```

From the accuracy results of two models I calculated two value for further work:
1. The accuracy of the train the subjects hasn't been used as train data (leaved_out)
2. The average of accuracy for every train (mean)

```python
SVM_leaved_out = np.diag(SVM_CrossVal)
SVM_mean = SVM_CrossVal.mean(axis=0)

CNN_leaved_out = np.diag(CNN_CrossVal)
CNN_mean = CNN_CrossVal.mean(axis=0)
```

I concatanated all available data to a single data frame:

```python
df = pd.concat([subjects, NASA_TLI], axis=1)
df["SVM_leaved_out"] = SVM_leaved_out
df["SVM_mean"] = SVM_mean
df["CNN_leaved_out"] = CNN_leaved_out
df["CNN_mean"] = CNN_mean
df.head()
```

```python
df.to_csv("data/summary_real.csv")
```
