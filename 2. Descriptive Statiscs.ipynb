```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
subjects = pd.read_csv('data/subjects.csv', index_col=0)
NASA_TLI = pd.read_csv('data/NASA_TaskLoadIndex.csv', index_col=0).T
SVM_CrossVal = pd.read_csv('data/SVM_CrossVal.csv', index_col=0)
CNN_CrossVal = pd.read_csv('data/CNN_CrossVal.csv', index_col=0)
```

```python
# Convert the gender column to a categorical variable
gender = {'male': 1,'female': 0}
subjects.Sex = [gender[item] for item in subjects.Sex]
```

```python
print("The distribution of the Age of the subjects is:")
_ = subjects.hist('Age', bins = 48)
plt.show()

print("The gender distribution of subjects:")
_ = subjects.hist('Sex',bins = 2)
plt.show()
```

```python
print("The distribution of the NASA-TLI scores is:")
_ = NASA_TLI.hist(bins=20, range = (1,20), figsize=(10, 8))
```

```python
print("The distribution of the SVM-CrossVal scores is:")
_ = sns.boxenplot(data=SVM_CrossVal, orient='h', palette='Set2')
```

```python
print("The distribution of the CNN-CrossVal scores is:")
_ = sns.boxenplot(data=CNN_CrossVal, orient='h', palette='Set2')
```
